```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
import seaborn as sns

df =sns.load_dataset('iris')
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = iris.target
```

```python
X
```

```python
y
```

```python
##train test plot
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel1 = DecisionTreeClassifier()
```

```python
treemodel1.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel1,filled=True)
```

```python
## Postprunnig
treemodel = DecisionTreeClassifier(max_depth=2)
treemodel.fit(X_train,y_train)
```

```python
## Post Preunning
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
y_pred = treemodel.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
## Pre Prunnig

parameters = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5],
    'max_features': ['sqrt', 'log2', None]  # Changed from 'auto'


}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel2 = DecisionTreeClassifier()
cv = GridSearchCV(treemodel2,param_grid=parameters,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_
```

```python
cv.predict(X_train)
```

```python
y_pred = cv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
