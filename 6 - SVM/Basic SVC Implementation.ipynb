## Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
import matplotlib.pyplot as plt

X, y = make_classification(n_samples=1000, n_features=2, n_classes=2,
                          n_clusters_per_class=2, n_redundant=0)

plt.figure(figsize=(8, 6))

plt.scatter(X[y == 0, 0], X[y == 0, 1], color='blue', alpha=0.5, label='Class 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='red', alpha=0.5, label='Class 1')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('2-Class Classification Data with 2 Features')
plt.legend()

plt.grid(True)
plt.show()
```

```python
sns.scatterplot(pd.DataFrame(X)[0],pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
## Prediction
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
## Prediction
y_pred1=rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(X_train,y_train)
## Prediction
y_pred2=polynomial.predict(X_test)
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
## Prediction
y_pred3=sigmoid.predict(X_test)
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid.intercept_
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
