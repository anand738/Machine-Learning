```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.neighbors import LocalOutlierFactor
```

```python
X, y = make_circles(n_samples=400, factor=0.3, noise=0.05)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
n_outliers = 20
outliers = np.random.uniform(low=-1.5, high=1.5, size=(n_outliers, 2))
X = np.vstack([X, outliers])
```

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)
```

```python
inliers = X[y_pred == 1]
outliers = X[y_pred == -1]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(inliers[:, 0], inliers[:, 1], c='blue', label='Inliers', s=50, alpha=0.6)
plt.scatter(outliers[:, 0], outliers[:, 1], c='red', label='Outliers', s=100, marker='x')
plt.title('Local Outlier Factor on make_circles Dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.grid(True)
plt.show()
```
